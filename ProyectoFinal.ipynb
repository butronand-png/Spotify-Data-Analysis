### ¿Cómo tratar los datos atípicos? 

* Debido a que no son errores y sólo son **valores extremos**, entonces se decide reemplazarlos con la media o con la mediana (apuntes Lourdes)

```python
import numpy as np
```

```python
plt.boxplot(atributos_nor)
```

```python
def tratar_outliers_con_mediana(df, columnas):
    """
    Identifica outliers con el método IQR y los reemplaza con la mediana.
    """
    df_tratado = df.copy()  # Copia para no modificar el original
    
    for col in columnas:
        # PASO 1: Calcular los cuartiles
        Q1 = df[col].quantile(0.25)  # 25% de los datos
        Q3 = df[col].quantile(0.75)  # 75% de los datos
        IQR = Q3 - Q1                # Rango entre Q1 y Q3
        
        # PASO 2: Definir qué es un outlier
        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR
        
        # PASO 3: Calcular la mediana (el valor del medio)
        mediana = df[col].median()
        
        # PASO 4: Identificar cuáles son outliers
        es_outlier_bajo = df[col] < limite_inferior
        es_outlier_alto = df[col] > limite_superior
        es_outlier = es_outlier_bajo | es_outlier_alto  # | significa "o"
        
        # PASO 5: Reemplazar outliers con la mediana
        df_tratado.loc[es_outlier, col] = mediana
    
    return df_tratado
```

```python
atributos_tratados = tratar_outliers_con_mediana(atributos_nor, atributos_nor.columns)
```

### Normalización de texto

```python
tokenizer = nltk.WordPunctTokenizer()
def normalizar_escalar(texto):
    #Eliminación de mayúsculas
    texto = texto.lower()
    #Eliminación de caracteres especiales, números 
    texto = re.sub("[^a-zA-Zúíóáéñ\s]", "", texto, re.I|re.A)
    #Se realiza proceso de tokenización
    tokens = tokenizer.tokenize(texto)
    #Se eliminan stop words de los tokens. Filteres tokens es una lista de palabras que no son 'stop words'
    filtered_tokens = [token for token in tokens if token not in stop_words]
    #Se reunen las palabras en filtered_tokens
    return " ".join(filtered_tokens)
```

### Lematización de texto

```python
def lematizar_escalar(oracion):
    #Se procesa el texto
    oracion = str(oracion)
    doc = nlp(oracion)
    lematizadas = [token.lemma_ for token in doc]
    return " ".join(lematizadas)
```

### Enraización de texto

```python
stemmer = SnowballStemmer("spanish")
def enraizar_escalar(oracion):
    oracion = str(oracion)
    palabras_con_raiz = []
    lista_oracion = []
    #Partir cada oración en la lista en palabras
    oracion = oracion.split() 
    # Hacer un ciclo for dentro de un ciclo for (cada palabra en cada oracion en el corpus debe ser analizada)
    for palabra in oracion:
        #Se crea una lista vacía ('lista_oracion') que luego será agregada a 'palabras_con_raiz'
        lista_oracion.append(stemmer.stem(palabra))
    corpus_raiz = " ".join(lista_oracion)
    palabras_con_raiz.append(lista_oracion)
    return corpus_raiz
```

### Funciones vectorizadas

```python
normalizar_vectorizado = np.vectorize(normalizar_escalar)
lematizar_vectorizado = np.vectorize(lematizar_escalar)
enraizar_vectorizado = np.vectorize(enraizar_escalar)
```

```python
palabras_artistas = []
# No usar el corpus normalizado
for artista in corpus_artistas:
    # Solo separar por comas (porque varios artistas están separados por comas)
    nombres = artista.split(',')
    for nombre in nombres:
        nombre = nombre.strip()  # Quitar espacios en blanco
        if nombre:  # Si no está vacío
            palabras_artistas.append(nombre)
```
